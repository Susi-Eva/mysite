### LOC

```python
temp = {}
locArr = []
for i in range(len(api_commit)):
    a = api_commit[i]['weeks']
    arrA = []
    for j in range(len(api_commit[i]['weeks'])):
        tempDict = api_commit[i]['weeks'][j]
        arr = []
        for key, value in tempDict.items():
            arr.append(value)
        arrA.append(arr)
    locArr.append(arrA)
```

```python
ArrAdd = []
ArrDel = []
for i in range(len(locArr)):
    tempArr= []
    tempArr2=[]
    for j in range(len(locArr[i])):
        tempArr.append(locArr[i][j][1])
        tempArr2.append(locArr[i][j][2])
        #print(j)
    
    ArrAdd.append(tempArr)
    ArrDel.append(tempArr2)
```

```python
countAdd = []
for i in range(len(ArrAdd)):
    count = 0
    for j in range(len(ArrAdd[i])):
        if ArrAdd[i][j] != 0:
            count+=1           
    countAdd.append(count)
```

```python
countDel = []
for i in range(len(ArrDel)):
    count = 0
    for j in range(len(ArrDel[i])):
        if ArrDel[i][j] != 0:
            count+=1           
    countDel.append(count)
```

```python
countAD = []
for i in range(len(countAdd)):
    result = countAdd[i]+countDel[i]
    countAD.append(result)
```

```python
LOC_dict = {}

for i in range(len(countAD)):
    developer = api_commit[i]['author']['login']
    LOC_dict[developer] = countAD[i]
```

```python
# make LOC to DATAFRAME

Frame_count_state_issues = pd.DataFrame(list(count_state_issues.items()), columns=['User Name','Issue Total'])
frame_LOC = pd.DataFrame(list(LOC_dict.items()), columns=['User Name', 'LOC Total'])
frame_LOC
```

### MERGE DATAFRAME

```python
df3 = frame_LOC
LOC_score = app.performance_LOC(df3)
LOC_score.head()
```

```python
# merger LOC and Commit Performance

mergeCLOC = pd.merge(commits_score,LOC_score, on='User Name')
```

```python
mergeIPRCLOC = pd.merge(pd.merge(issue_score, pulls_score, on='User Name'), mergeCLOC, on='User Name', how='outer')
mergeIPRCLOC.fillna(0,inplace = True)
mergeIPRCLOC.head()
```

```python
# jumlahkan nilai issue, pull request, commit dan loc

issue = []
pull = []
commit = []
loc = []

for i in range(len(mergeIPRCLOC['User Name'])):
    iss = mergeIPRCLOC['issue_score'][i]
    pll = mergeIPRCLOC['pull_score'][i]
    cmm = mergeIPRCLOC['commit_score'][i]
    lc = mergeIPRCLOC['LOC_score'][i]
    issue.append(iss)
    pull.append(pll)
    commit.append(cmm)
    loc.append(lc)
    
totalState = []
for j in range(len(issue)):
    result = issue[j] + pull[j] + commit[j] + loc[j]
    totalState.append(result)
    
```

```python
mergeIPRCLOC['Total Performance of Developer'] = totalState
mergeIPRCLOC
```

```python
df0 = issue_score
df1 = pulls_score
df2 = commits_score
df3 = LOC_score

Performance_score = app.total_performance(df0, df1, df2, df3)
Performance_score
```
